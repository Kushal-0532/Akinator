Calculating Info gain for each feature and showing that!


```python
# for each feature in data
import numpy as np
info_gain = {}  # Dictionary to store feature and its info gain

for feature in X:
    yes_count = np.sum(X[feature] == 1)
    no_count = np.sum(X[feature] == 0)

    p_yes = yes_count / (yes_count + no_count)
    p_no = no_count / (yes_count + no_count)

    h_yes = -p_yes * np.log2(p_yes) if p_yes > 0 else 0
    h_no = -p_no * np.log2(p_no) if p_no > 0 else 0

    h_split = p_yes * h_yes + p_no * h_no
    info_gain[feature] = h_split

# Step 2: Sort features by info gain
sorted_features = sorted(info_gain.items(), key=lambda x: x[1], reverse=True)

# Step 3: Print sorted features
for feature, ig in sorted_features:
    print(f"{feature}: {ig}")


```

```python
# remove features with h_split value of <0.05

for feature, ig in sorted_features:
  if ig < 0.07:
    X = X.drop(feature, axis=1)

print("Features: ")
for feature in X:
  print(feature)

from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=3)  # You can adjust n_neighbors as needed
model.fit(X, y)
```

```python
import random

# make a function that just puts feature key to value as "does the character have attribure {feature} in a dictionary called questions"
questions = {}
print("These are the values from 0-5 for hair color: ",'Black Hair',' Blonde Hair',' Blue Hair',' Brown Hair',' Green Hair',' Grey Hair')
print("These are the values from 0-2 for gender: ",'Female ','Male',' Neutral')
for feature in X:
  questions[feature] = "Does the character have attribute {0}?".format(feature)

sorted_feature_names = [feature for feature, _ in sorted_features]
# sorting questions's features like sorted_features
questions = {k: v for k, v in sorted(questions.items(), key=lambda item: sorted_feature_names.index(item[0]))}
input_from_user = {}
for feature,question in questions.items():
  print("{0}: ".format(question))


  # put a random 1 or 0 here
  input_from_user[feature] = input() #------------------------------------>
  #input_from_user[feature] = random.randint(0,1)

input_array = np.zeros((1, len(X.columns)))
for i, feature in enumerate(X.columns):
    if feature in input_from_user:
        input_array[0, i] = input_from_user[feature]

# Make prediction
predicted_character = model.predict(input_array)[0]
predicted_probs = model.predict_proba(input_array)[0]
max_prob_index = np.argmax(predicted_probs)
probability = predicted_probs[max_prob_index] * 100

# Get top 3 most likely characters
top_indices = np.argsort(predicted_probs)[-3:][::-1]
class_labels = list(model.classes_)

print("\nPrediction Results:")
print(f"I think your character is: {predicted_character} (Confidence: {probability:.2f}%)")

print("\nTop 3 most likely characters:")
for i, idx in enumerate(top_indices):
    char_name = class_labels[idx]
    char_prob = predicted_probs[idx] * 100
    print(f"{i+1}. {char_name}: {char_prob:.2f}%")
```
